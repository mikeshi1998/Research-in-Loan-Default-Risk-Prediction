```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import datetime
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
pd.set_option('display.max_columns', None)  # this will output all columns when display a pandas dataframe
```

```python
df=pd.read_csv('2012-2013_cleaned2.csv', index_col=0)
```

```python
df.head()
```

```python
y=df['loan_status']
df.drop(columns=['emp_title','issue_d','desc','title','zip_code','loan_status'], inplace=True)
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.3)
scaler=StandardScaler()
scaler.fit(X_train)
scaled_train=scaler.transform(X_train)
scaled_test=scaler.transform(X_test)
```

```python
logmodel = LogisticRegression()
logmodel.fit(scaled_train,y_train)
predictions = logmodel.predict(scaled_test)
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
print(logmodel.score(scaled_test, y_test))
```

```python
param_grid={'penalty': ['l1','l2','none'], 'C':[0.01,0.1,1,5,10,50,100,500,1000],
            'class_weight': [None, 'balanced'], 'solver': ['liblinear','newton-cg','lbfgs','sag','saga']}
```

```python
grid = GridSearchCV(LogisticRegression(), param_grid, refit=True, error_score=0)
grid.fit(scaled_train,y_train)
```

```python
grid.best_params_
```

```python
grid_predictions = grid.predict(scaled_test)
print(classification_report(y_test,grid_predictions))
print(confusion_matrix(y_test,grid_predictions))
print(grid.score(scaled_test, y_test))
```
